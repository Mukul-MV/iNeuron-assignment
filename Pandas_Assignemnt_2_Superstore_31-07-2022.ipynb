**15 . try to find out the 10th most loss making product catagory**

```python
# Thier are only 3 Product Category , so here I have sort according to highest loss macking to lowest 
'''logic == find ratio of negative profit(Loss) by positive profit(Profit) and whichever ratio is higher that category 
will be consider as a most loss making category over profit generation. 
'''
```

```python
#stores Loss
df_profit = df_store[df_store['Profit']<0].groupby('Product Category')[['Profit']].sum()
```

```python
df_profit.rename(columns = {'Profit':'Loss'},inplace=True)
```

```python
#stores profit only
df_profit['Profit'] = df_store[df_store['Profit']>0].groupby('Product Category')[['Profit']].sum()
```

```python
#stores total profit
df_profit['Total_profit'] = df_profit.sum(axis = 1)
```

```python
df_profit['loss_profit_ratio'] = df_profit['Loss']/df_profit['Profit']
```

```python
df_profit.sort_values(by = ['loss_profit_ratio'])
```
