**8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset**

```python
req_calorie_burn_per_day   = 2500
```

```python
# count healthy days of each person if their calories burn is greater than 2500
df_health = df[df['Calories'] >= 2500].groupby(['Id'])[['Calories']].count()
```

```python
df_health.rename(columns = {'Calories':'healty_days'},inplace=True)
```

```python
df_health['Total calories'] = df.groupby(['Id'])['Calories'].sum()
```

```python
# total loggin days per person
df_health['Total days logged'] = df.groupby('Id')['ActivityDate'].count()
```

```python
df_health['Avg_Calories'] = df_health['Total calories']/df_health['Total days logged']
```

```python
# health index ==  healthy days / Total logged days

df_health['health_index'] = df_health['healty_days']/df_health['Total days logged']
```

```python
df_health
```

```python
df_health.describe()
```

```python
# healthy person ==  either health_index>0.6 or avg_calories > 2500

df_health[(df_health['health_index']>=0.6) | (df_health['Avg_Calories']>=2500)].sort_values(['health_index','Avg_Calories'],ascending=False)
```
